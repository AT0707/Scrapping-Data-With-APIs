```python
calc_request = Ytube.search().list(part = "snippet", maxResults = 25, q = 'calculus', type = 'videos') 
calc_results = calc_request.execute()
```

```python
calc_results
```

```python
title=[]
videoid=[]
videocontent=[]
videochannel=[]
channelid=[]
videopostdate=[]
channeltitle=[]
rep_len=len(calc_results['items'])
for i in range(rep_len):
    if calc_results['items'][i]['id']['kind']=='youtube#video':
        channeltitle.append(calc_results['items'][i]['snippet']['channelTitle'])
        title.append(calc_results['items'][i]['snippet']['title'])
        videoid.append(calc_results['items'][i]['id']['videoId'])
        videochannel.append(calc_results['items'][i]['snippet']['channelTitle'])
        channelid.append(calc_results['items'][i]['snippet']['channelId'])
        videocontent.append(calc_results['items'][i]['snippet']['description'])
        videopostdate.append(calc_results['items'][i]['snippet']['publishedAt'])
```

```python
likes,views,favs,comments=[],[],[],[]
for idt in videoid:
    a=Ytube.videos().list(part="statistics",id=idt).execute()
    likes.append(a['items'][0]['statistics']['likeCount'])
    views.append(a['items'][0]['statistics']['viewCount'])
    comments.append(a['items'][0]['statistics']['commentCount'])
    favs.append(a['items'][0]['statistics']['favoriteCount'])
```

```python
title
```

```python

Calc_data=pd.DataFrame({'Views': views, 'Likes':likes, 'Number of comments': comments, 'Titles': title, 'Channel Titles': channeltitle, 'Channel ID':channelid})
```
